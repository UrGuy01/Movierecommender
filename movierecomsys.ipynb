```python
#since genres are not in the required format we will create a function to recreate the format....
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
#for instance we dont need the above format for our app...we need in the format["action",'adventure','fantasy',....etc]]
import ast 
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
        return L
    
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head(5)
#problem solved for genres section...
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head(2)
#keywords now are in presentable format...
```

```python
movies['cast'][0]
```

```python
#now for cast...we need only the important people to promote the film to user...(for example, top three characters...)
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
    
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies['cast']
```

```python
movies.head(3)
#now cast is in presentable format ...
```

```python
#now same for the crew
movies['crew'][2]
# I only care for job title director
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            L.append(i["name"])
            break  # Assuming you only want to append the first director
    return L
```

```python
movies['crew'] = movies['crew'].apply(lambda x: fetch_director(x) if x else [])
```

```python
movies.head(3)
```

```python
movies['overview'][0]
#since the overview is in string form and we want it to be in list form so we can concatenate with others to form  tags...thus,
movies['overview'].apply(lambda x:x.split())
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
#cleaning spaces-transformation
movies['crew'].head(1)
#so here james cameroon shld be in the form jamescameroon
movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])




```

```python
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x] if isinstance(x, list) else None)

movies['keywords'][0]
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x] if isinstance(x, list) else None)
movies['genres'][0]
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
movies.head(2)

```

```python
new_df=movies[['title','movie_id','tags']]
new_df
```

```python
# Assuming new_df is your DataFrame containing the data you provided

# Convert lists in 'tags' column to strings
new_df['tags'] = new_df['tags'].apply(lambda x: ' '.join(x) if isinstance(x, list) else x)


```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
        return " ".join(y)
    
```

```python
#example of stemming....
#if i have words [like love,loving,loved,lover]
#it will return love(reducing the dimensionality of the dataset)
['loving','loved','lover']
['love','love','love']
ps.stem('loving')
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[1]
#it basically gave me the comparison values of movie 1 to all the other movies in the dataset in the values between 0 to 1....1 being the most similar type of movie ....
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]#this will give the the similarity of first movie with respect to all the other movie according to their index values
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]]['title'])
    
    

    
    
```

```python
recommend('John Carter')
```
